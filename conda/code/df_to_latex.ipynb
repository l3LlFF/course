```python
import pandas as pd
import numpy as np
```

### experiment 1

```python
df_1 = pd.read_excel('exp_1/K.xlsx')
df_1
```

```python
df_2 = pd.read_excel('exp_1/rho.xlsx')
df_2
```

```python
df_all = pd.concat([df_1, df_2], axis=1)
df_all
```

```python
R = 4
col = 3
k = 1
for i in range(len(df_all)):
    s = f'{k} &'
    for j in range(col):
        s += f"{round(df_all.iloc[i, 1 + j], R)} & "
        s += f"{round(np.log10(df_all.iloc[i, 1 + j]), R)} & "
        s += f"{round(df_all.iloc[i, 2 + col + j], R)} & "
    k += 1
    print(s[:-2] + "\\\\ \\hline")
```
